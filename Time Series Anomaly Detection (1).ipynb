This function, named `overview`, takes a Pandas DataFrame (`df`) and an optional timestamp column name (`timestamp_col`). It prints the count of null values and data types of each column in the DataFrame. If a timestamp column is specified, it also displays the date range, start and end dates, and the total number of days covered by the timestamps in the specified column.

```python
def overview(df: pd.DataFrame, timestamp_col: str = None) -> None:
    print('Null Count:\n', df.isnull().sum(),'\n')
    print('Data Types:\n', df.dtypes)
    
    if timestamp_col is not None:
        print('\nDate Range:\n\nStart:\t',df[timestamp_col].min())
        print('End:\t',df[timestamp_col].max())
        print('Days:\t',(df[timestamp_col].max() - df[timestamp_col].min()))
```

1. **Hourly Resampling:**
   - `df_hourly` is created by setting the timestamp column as the index, resampling the data at hourly intervals using `resample('H')`, calculating the mean demand within each hour using `mean()`, and then resetting the index.

2. **Daily Resampling:**
   - `df_daily` is created similarly but resampling at daily intervals using `resample('D')`. This results in the mean demand for each day.

3. **Weekly Resampling:**
   - `df_weekly` follows the same pattern but with weekly resampling using `resample('W')`, giving the mean demand for each week.

These resampled DataFrames provide aggregated views of the original data at different temporal granularities (hourly, daily, and weekly).
```python
# A variety of resamples which I may or may not use
df_hourly = df.set_index('timestamp').resample('H').mean().reset_index()
df_daily = df.set_index('timestamp').resample('D').mean().reset_index()
df_weekly = df.set_index('timestamp').resample('W').mean().reset_index()
print(df_hourly.head(5))
print(df_daily.head(5))
print(df_weekly.head(5))
```

Creating new features in two resampled DataFrames (`df_hourly` and `df_daily`). The loop iterates through each DataFrame and adds several time-related and derived features:

1. **Loop through DataFrames:**
   - The loop iterates through each DataFrame (`df_hourly` and `df_daily`).

2. **Date and Time Features:**
   - `Weekday`: Creates a new categorical feature representing the day of the week based on the timestamp. The day names are ordered from Monday to Sunday.
   - `Hour`: Extracts the hour from the timestamp.
   - `Day`: Extracts the day of the week (numeric representation) from the timestamp.
   - `Month`: Extracts the month from the timestamp.
   - `Year`: Extracts the year from the timestamp.
   - `Month_day`: Extracts the day of the month from the timestamp.

3. **Lag Feature:**
   - `Lag`: Creates a lag feature by shifting the 'value' column by one time step. This represents the value from the previous timestamp.

4. **Rolling Mean Feature:**
   - `Rolling_Mean`: Calculates a rolling mean with a window size of 7 and a minimum number of periods set to 1. This creates a feature representing the average demand over a 7-day window, updating with each timestamp.

5. **Data Cleaning:**
   - `DataFrame = DataFrame.dropna()`: Removes rows with missing values, ensuring the new features are computed on a clean dataset.

These new features can be valuable for time-series analysis, helping to capture temporal patterns and relationships in the data. The loop efficiently adds these features to both the hourly and daily resampled DataFrames.

```python
for DataFrame in [df_hourly, df_daily]:
    DataFrame['Weekday'] = (pd.Categorical(DataFrame['timestamp'].dt.strftime('%A'),
                                           categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday', 'Saturday', 'Sunday'])
                           )
    DataFrame['Hour'] = DataFrame['timestamp'].dt.hour
    DataFrame['Day'] = DataFrame['timestamp'].dt.weekday
    DataFrame['Month'] = DataFrame['timestamp'].dt.month
    DataFrame['Year'] = DataFrame['timestamp'].dt.year
    DataFrame['Month_day'] = DataFrame['timestamp'].dt.day
    DataFrame['Lag'] = DataFrame['value'].shift(1)
    DataFrame['Rolling_Mean'] = DataFrame['value'].rolling(7, min_periods=1).mean()
    DataFrame = DataFrame.dropna()
```

This function, `run_isolation_forest`, applies the Isolation Forest algorithm for anomaly detection to the input DataFrame `model_data`:

1. **Parameters:**
   - `model_data: pd.DataFrame`: The input DataFrame containing the features for anomaly detection.
   - `contamination=0.005`: The proportion of outliers or anomalies expected in the data (default is 0.005 or 0.5%).
   - `n_estimators=200`: The number of base estimators in the ensemble (default is 200).
   - `max_samples=0.7`: The number of samples to draw from the input data to train each base estimator, expressed as a fraction of the total number of samples (default is 0.7 or 70%).

2. **Isolation Forest Initialization:**
   - `IF = IsolationForest(random_state=0, contamination=contamination, n_estimators=n_estimators, max_samples=max_samples)`: Initializes an instance of the Isolation Forest with specified hyperparameters.

3. **Model Fitting:**
   - `IF.fit(model_data)`: Fits the Isolation Forest model to the input data (`model_data`).

4. **Prediction and Scoring:**
   - `output = pd.Series(IF.predict(model_data)).apply(lambda x: 1 if x == -1 else 0)`: Predicts anomalies based on the Isolation Forest model. The output is a binary series where 1 indicates an anomaly.
   - `score = IF.decision_function(model_data)`: Computes the anomaly score for each sample in the input data.

5. **Output:**
   - The function returns a tuple containing the anomaly predictions (`output`) and the anomaly scores (`score`).

This function is designed to identify anomalies in the input data using the Isolation Forest algorithm and provide both binary anomaly predictions and anomaly scores for further analysis. The hyperparameters such as `contamination`, `n_estimators`, and `max_samples` can be adjusted based on specific use cases and data characteristics.

```python
def run_isolation_forest(model_data: pd.DataFrame, contamination=0.005, n_estimators=200, max_samples=0.7) -> pd.DataFrame:
    
    IF = (IsolationForest(random_state=0,
                          contamination=contamination,
                          n_estimators=n_estimators,
                          max_samples=max_samples)
         )
    
    IF.fit(model_data)
    
    output = pd.Series(IF.predict(model_data)).apply(lambda x: 1 if x == -1 else 0)
    
    score = IF.decision_function(model_data)
    
    return output, score
```
